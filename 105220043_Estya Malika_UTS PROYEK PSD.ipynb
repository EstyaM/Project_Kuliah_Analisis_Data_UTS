```python
#Menggambarkan variasi dalam distribusi data
print(df['produksi_sayuran_buah'].describe())
sns.distplot(df['produksi_sayuran_buah'], color='g', bins=25, hist_kws={'alpha': 0.4});
```

```python
#Menampilkan data dalam bentuk boxplot agar mudah mengetahui outlier dan lainnya
sns.boxplot(data=df[['produksi_sayuran_buah']])
```

```python
#Mencari Outlier 
def iqr_outlier(data,alpha):
  q75, q25 = np.percentile(data, [75 ,25])
  iqr = q75 - q25
  upperbound=q75+(iqr*alpha)
  lowerbound=q25-(iqr*alpha)
  return upperbound,lowerbound
```

```python
#Mencari nilai Upperbound dan Lowerbound
u_produksi_sayuran_buah,l_produksi_sayuran_buah=iqr_outlier(df['produksi_sayuran_buah'],alpha=1.5)
print(u_produksi_sayuran_buah,l_produksi_sayuran_buah)
```

```python
#Menampilkan data yang berada diantara upperbound dan lowerbound
df[
    (df['produksi_sayuran_buah']>u_produksi_sayuran_buah) |
    (df['produksi_sayuran_buah']<l_produksi_sayuran_buah)
]
```
